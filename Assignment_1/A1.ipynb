```python
# Count the number of zero values in each column
zero_counts = (df == 0).sum()
print("Zero values in each column:")
print(zero_counts)
```

```python
# Select a specific row (e.g., row 5)
print(df.iloc[4])

# Select a specific column (e.g., 'Age')
print(df['age'])

# Select a subset of rows and columns
print(df.loc[0:10, ['age', 'chol']])
```

```python
# Filter rows where 'Age' > 50
filtered_df = df[df['age'] > 50]
print(filtered_df)
```

```python
# Sort by 'Age' in ascending order
sorted_df = df.sort_values(by='age', ascending=True)
print(sorted_df)
```

```python
# Select a single column
age_column = df['age']

# Select multiple columns
selected_columns = df[['age', 'sex', 'chol']]

# Select rows using iloc (by index)
first_5_rows = df.iloc[:5]

# Select rows using loc (by condition)
high_chol = df[df['chol'] > 200]  # Filter rows where cholesterol is greater than 200

print(selected_columns.head())
print(first_5_rows)
print(high_chol)
```

```python
# Sort by a single column (ascending)
sorted_df = df.sort_values(by='age')

# Sort by multiple columns (descending)
sorted_df_desc = df.sort_values(by=['age', 'chol'], ascending=[False, True])

print(sorted_df.head())
print(sorted_df_desc.head())
```

```python
# Count unique values in each column
print(df.nunique())

# Count unique values in a specific column
print(df['sex'].value_counts())

```

```python
# Display column names and their data types
for col in df.columns:
    print(f"Column: {col}, Data Type: {df[col].dtype}")
```

```python
# Convert 'age' from int to float
df['age'] = df['age'].astype(float)

# Convert 'sex' from integer (0,1) to string (Male, Female)
df['sex'] = df['sex'].map({1: 'Male', 0: 'Female'})

# Convert 'chol' to categorical (High, Normal)
df['chol_category'] = pd.cut(df['chol'], bins=[0, 200, 600], labels=['Normal', 'High'])

print(df.dtypes)
print(df.head())
```
